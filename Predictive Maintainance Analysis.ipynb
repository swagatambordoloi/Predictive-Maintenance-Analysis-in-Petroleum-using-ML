```python
#1. Clean missing values from critical tracking sensors to avoid plotting drops
severe_slugging_master = severe_slugging_master.dropna(subset=["P-TPT", "P-PDG"])
```

```python
# 2. Apply Universal Anomaly Mask (Optimized with fillna to prevent NaN traps)
anomaly_mask = severe_slugging_master['class'].fillna(0) != 0
anomaly_data = severe_slugging_master[anomaly_mask]
```

```python
# 3. Extract the exact transition and boundary points
start_idx = anomaly_data.index.min()
end_idx = anomaly_data.index.max()

print(f"✔ Anomaly Transition Point Found at Row Index: {start_idx}")
print(f"✔ Severe Slugging Active Windows Span: Row {start_idx} to Row {end_idx}")

# 4. Create a precise 3,000-row zoom window around the onset (500 rows before, 2500 rows after)
zoom_start = max(0, start_idx - 500)
zoom_end = min(severe_slugging_master.index.max(), start_idx + 2500)
zoomed_slugging = severe_slugging_master.loc[zoom_start:zoom_end]
```

```python
#PLOT THE OSCILLATING WAVE SIGNATURES
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 9), sharex=True)
# Plot Downhole Pressure (P-PDG)
ax1.plot(zoomed_slugging.index, zoomed_slugging['P-PDG'], color='#d90429', linewidth=1.5, label='Downhole Pressure (P-PDG)')
ax1.axvline(x=start_idx, color='#2b2d42', linestyle='--', linewidth=2, label=f'Slugging Onset (Row {start_idx})')
ax1.set_title("Zoomed Analysis: Sensor Signature Changes at Slugging Transition Point", fontsize=14, fontweight='bold')
ax1.set_ylabel("P-PDG (Downhole Pressure)", fontsize=11)
ax1.legend(loc="upper left")
ax1.grid(True, linestyle=':', alpha=0.6)

# Plot Wellhead Pressure (P-TPT)
ax2.plot(zoomed_slugging.index, zoomed_slugging['P-TPT'], color='#0077b6', linewidth=1.5, label='Wellhead Pressure (P-TPT)')
ax2.axvline(x=start_idx, color='#2b2d42', linestyle='--', linewidth=2)
ax2.set_xlabel("Dataframe Row Index Timeline", fontsize=11)
ax2.set_ylabel("P-TPT (Wellhead Pressure)", fontsize=11)
ax2.legend(loc="upper left")
ax2.grid(True, linestyle=':', alpha=0.6)

plt.tight_layout()
plt.savefig('slugging_transition_plot.png')
plt.show()
```

```python
#STEP 3: QUANTIFY AMPLITUDE & VARIANCE PROFILE
# Isolate specific window boundaries for statistical comparison
healthy_window = severe_slugging_master.loc[0:start_idx-1]
active_window = severe_slugging_master.loc[start_idx:end_idx]

print("\n--- QUANTIFIED SENSOR VARIANCE PROFILE ---")
print(f"Healthy Window P-PDG  -> Min: {healthy_window['P-PDG'].min():,.2f} | Max: {healthy_window['P-PDG'].max():,.2f} | Std Dev: {healthy_window['P-PDG'].std():,.2f}")
print(f"Slugging Window P-PDG -> Min: {active_window['P-PDG'].min():,.2f} | Max: {active_window['P-PDG'].max():,.2f} | Std Dev: {active_window['P-PDG'].std():,.2f}")
print("-" * 75)
print(f"Healthy Window P-TPT  -> Min: {healthy_window['P-TPT'].min():,.2f} | Max: {healthy_window['P-TPT'].max():,.2f} | Std Dev: {healthy_window['P-TPT'].std():,.2f}")
print(f"Slugging Window P-TPT -> Min: {active_window['P-TPT'].min():,.2f} | Max: {active_window['P-TPT'].max():,.2f} | Std Dev: {active_window['P-TPT'].std():,.2f}")
```

```python
import pandas as pd
normal_master=pd.read_csv(r"C:\Users\simpi\Downloads\normal_master.csv")
slugging_master=pd.read_csv(r"C:\Users\simpi\Downloads\Slugging-COMBINED.csv")
hydrates_master=pd.read_csv(r"C:\Users\simpi\Downloads\hydrates_master.csv")                        
```

```python
healthy_sample=normal_master.sample(n=3000,random_state=42)
slugging_sample=slugging_master[slugging_master['class']==3].sample(n=3000,random_state=42)
hydrates_sample=hydrates_master[hydrates_master['class'].isin([108,8])].sample(n=3000,random_state=42)
```

```python
healthy_sample['Asset_State']='Healthy'
slugging_sample['Asset_State']='Slugging'
hydrates_sample['Asset_State']='Hydrates'
```

```python
master_df=pd.concat([healthy_sample,slugging_sample,hydrates_sample],ignore_index=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=master_df,
    x='P-PDG',
    y='P-TPT',
    hue='Asset_State',
    alpha=0.6
)
plt.title('Multi-Fault Scatter Overlay')
plt.xlabel('Downhole Pressure (P-PDG)')
plt.ylabel('Wellhead Pressure (P-TPT)')
plt.show()
```

```python
import plotly.express as px

# Create the interactive 3D scatter plot
fig = px.scatter_3d(
    master_df, 
    x='P-PDG', 
    y='P-TPT', 
    z='T-TPT',
    color='Asset_State',  # Distinct colors for Healthy, Slugging, and Hydrates
    opacity=0.6,          # Helps see through dense clusters of data points
    title='3D Thermodynamic State Space: P-PDG vs P-TPT vs T-TPT',
    labels={
        'P-PDG': 'Downhole Pressure (P-PDG)',
        'P-TPT': 'Wellhead Pressure (P-TPT)',
        'T-TPT': 'Temperature (T-TPT)'
    }
)

# Update marker size for better visual clarity
fig.update_traces(marker=dict(size=4))

# Adjust the layout to make it spacious
fig.update_layout(
    margin=dict(l=0, r=0, b=0, t=40),
    legend=dict(title_text='Asset Operational State')
)

# Render the plot in your Jupyter Notebook
fig.show()
```
